```python
#sekarang kita ingin medapatkan semua buku dalam 1 halaman yang artinya kita harus medapat semua url buku dalam 1 halaman tersebut
url=requests.get("https://books.toscrape.com/")
soup=BeautifulSoup(url.text,'lxml')

url_buku=soup.find_all('article',class_='product_pod')
list_buku=[]
for x in url_buku:
  per_buku=f"https://books.toscrape.com/catalogue/{x.h3.a['href']}"
  list_buku.append(per_buku)
list_buku
```

```python
#kalo kita mau mengambil smua url buku dari tiap halaman maka kita butuhkan loop dalam loop
# Mendefinisikan tempat yang menampung semua url halaman
url_pages = []

# Mendefinisikan halaman pertama adalah 1
page = 1
init_page = f"https://books.toscrape.com/catalogue/page-{page}.html"

# Looping pengecekan status halaman
while requests.get(init_page).status_code == 200:
  url_pages.append(init_page)
  page += 1
  init_page = f"https://books.toscrape.com/catalogue/page-{page}.html"
  
```

```python
list_buku=[]
for x in url_pages:
  url=requests.get(x)
  url_buku=soup.find_all('article',class_='product_pod')
  for y in url_buku:
    per_buku=f"https://books.toscrape.com/{y.h3.a['href']}"
    list_buku.append(per_buku)
list_buku
```

```python
def Data(url_buku):
  url=requests.get(url_buku)
  soup = BeautifulSoup(url.text,'lxml')
  judul = soup.find("div", class_ = "col-sm-6 product_main").h1.text
  harga = soup.find("p", class_ = "price_color").text
  stok = soup.find("p", class_ = "instock availability").text
  stok = stok.strip()
  url_image = soup.find("article", class_ = "product_page").img.get('src')
  url_image = "https://books.toscrape.com/" + '/'.join(url_image.split('/')[2:])
  rating = soup.find("p", class_ = re.compile("star-rating")).get("class")[1]
  kategori = soup.find("ul", class_ = "breadcrumb").findAll("li")[2].a.text
  deskripsi = soup.find("article", class_ = "product_page").findAll("p")[-1].text.strip()
  return {'Judul_Buku': judul, 'url_buku': url_buku, 'Harga_Buku': harga, 'Stok': stok, 'url_image': url_image, 'rating': rating, 'kategori': kategori, 'deskripsi': deskripsi}
```

```python
perpustakaan=pd.DataFrame()
for url_buku in list_buku:
  per_buku=Data(url_buku)
  perpustakaan = perpustakaan.append(per_buku, ignore_index=True)
```
